```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
file_path = r'C:\Users\Andy\Documents\CodingDojo\Tareas_de_Clase\Data\Sample Media Spend Data.csv'
df = pd.read_csv(file_path)

# Definir los rangos de ventas
bins = [0, 100000, 500000, 1000000, df['Sales'].max()]
labels = ['0-100K', '100K-500K', '500K-1M', '1M+']

# Asignar cada valor de ventas a un rango
df['Sales_Range'] = pd.cut(df['Sales'], bins=bins, labels=labels, include_lowest=True)

# Agrupar por semana (u otro período) y por rango de ventas
sales_by_range = df.groupby(['Calendar_Week', 'Sales_Range'])['Sales'].sum().unstack()

# Graficar
plt.figure(figsize=(12, 6))
sales_by_range.plot(kind='line', stacked=True, ax=plt.gca())

# Título y etiquetas
plt.title("Ventas Agrupadas por Rangos a lo Largo del Tiempo", fontsize=16)
plt.xlabel("Fecha", fontsize=14)
plt.ylabel("Ventas (agrupadas)", fontsize=14)

# Mejorar la visualización de las fechas en el eje x
plt.xticks(rotation=45)
plt.grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
