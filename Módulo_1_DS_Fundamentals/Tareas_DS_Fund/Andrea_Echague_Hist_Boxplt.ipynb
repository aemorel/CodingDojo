```python
import pandas as pd

# Cargar los archivos CSV
df_2015 = pd.read_csv(r'C:\Users\Andy\Documents\CodingDojo\Tareas_DS_Fund\Data\2015.csv')
df_2016 = pd.read_csv(r'C:\Users\Andy\Documents\CodingDojo\Tareas_DS_Fund\Data\2016.csv')
df_2017 = pd.read_csv(r'C:\Users\Andy\Documents\CodingDojo\Tareas_DS_Fund\Data\2017.csv')
df_2018 = pd.read_csv(r'C:\Users\Andy\Documents\CodingDojo\Tareas_DS_Fund\Data\2018.csv')
df_2019 = pd.read_csv(r'C:\Users\Andy\Documents\CodingDojo\Tareas_DS_Fund\Data\2019.csv')

# Revisar el número de filas, columnas y primeros datos de cada dataset
datasets = {'df_2015': df_2015, 'df_2016': df_2016, 'df_2017': df_2017, 'df_2018': df_2018, 'df_2019': df_2019}

for name, df in datasets.items():
    print(f"\n--- {name} ---")
    print(f"Shape: {df.shape}")  # Muestra el número de filas y columnas
    print(f"Columnas: {df.columns.tolist()}")  # Muestra los nombres de las columnas
    print(df.head(), "\n")  # Muestra las primeras 5 filas para obtener una idea de cómo son los datos
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurar que todas las columnas tengan nombres consistentes en cada dataset
df_2015.rename(columns={'Country': 'Country or region', 'Happiness Score': 'Score', 'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)': 'Healthy life expectancy', 'Freedom': 'Freedom to make life choices'}, inplace=True)
df_2016.rename(columns={'Country': 'Country or region', 'Happiness Score': 'Score', 'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)': 'Healthy life expectancy', 'Freedom': 'Freedom to make life choices'}, inplace=True)
df_2017.rename(columns={'Country': 'Country or region', 'Happiness.Score': 'Score', 'Economy..GDP.per.Capita.': 'GDP per capita', 'Health..Life.Expectancy.': 'Healthy life expectancy', 'Freedom': 'Freedom to make life choices'}, inplace=True)

# Renombrar los datasets de 2018 y 2019 también
df_2018.rename(columns={'Country or region': 'Country or region'}, inplace=True)
df_2019.rename(columns={'Country or region': 'Country or region'}, inplace=True)

# Concatenar los datasets
df_all = pd.concat([df_2015[['Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']],
    df_2016[['Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']],
    df_2017[['Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']],
    df_2018[['Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']],
    df_2019[['Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']]])

# Definir colores diferentes para histogramas y boxplots
colors_hist = ['skyblue', 'lightcoral', 'lightgreen', 'lightpink']
colors_box = ['darkblue', 'darkred', 'darkgreen', 'darkorange']

# Variables de interés
variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

# Crear gráficos con colores diferentes para histogramas y boxplots
for i, variable in enumerate(variables):
    plt.figure(figsize=(12, 6))
    
    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(df_all[variable], bins=20, kde=True, color=colors_hist[i])
    plt.title(f'Histograma de {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frecuencia')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(df_all[variable], color=colors_box[i])
    plt.title(f'Boxplot de {variable}')
    plt.xlabel(variable)
    
    plt.tight_layout()
    plt.show()
```

### **Distribución del Puntaje de Felicidad (Score):**
- **Histograma**: La mayoría de los países tienen un puntaje de felicidad que se concentra entre 4 y 7 puntos. Esto indica que la felicidad global está centrada alrededor de estos valores, con muy pocos países reportando puntajes extremadamente bajos o extremadamente altos.
- **Boxplot**: El boxplot revela algunos valores atípicos, tanto en el extremo superior como en el inferior. Esto sugiere que hay países con niveles de felicidad significativamente diferentes (mucho más altos o más bajos) en comparación con la mayoría.


### **Distribución del PIB per Cápita (GDP per capita):**
- **Histograma**: La distribución muestra un sesgo hacia la derecha, lo que indica que la mayoría de los países tienen un PIB per cápita bajo, mientras que unos pocos países tienen valores extremadamente altos. Esto es típico de la disparidad económica global, donde unos pocos países tienen una economía mucho más fuerte en comparación con la mayoría.
- **Boxplot**: El boxplot confirma la presencia de valores atípicos en la cola derecha. Estos países tienen un PIB per cápita mucho más alto que el promedio mundial, indicando una gran desigualdad en términos de riqueza entre países.


### **Distribución de la Expectativa de Vida Saludable (Healthy life expectancy):**
- **Histograma**: La distribución es relativamente simétrica, con un pico alrededor de los 65-70 años. Esto sugiere que la mayoría de los países tienen una expectativa de vida saludable moderada.
- **Boxplot**: Sin embargo, el boxplot muestra varios valores atípicos, especialmente en los extremos inferior y superior. Esto indica que algunos países tienen una expectativa de vida saludable considerablemente más baja o más alta que el promedio, posiblemente debido a factores como acceso a la atención médica, calidad de vida, o diferencias socioeconómicas.


### **Distribución de la Libertad para Tomar Decisiones de Vida (Freedom to make life choices):**
- **Histograma**: Similar al PIB per cápita, la distribución muestra un sesgo hacia la derecha, lo que indica que la libertad para tomar decisiones de vida es moderada o baja en la mayoría de los países, mientras que unos pocos países gozan de una mayor libertad.
- **Boxplot**: El boxplot muestra varios valores atípicos, lo que sugiere que hay algunos países donde la libertad para tomar decisiones de vida es significativamente diferente (ya sea más alta o más baja) en comparación con el promedio global.
