### **Código completo**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('tips')

# Crear una figura
fig, ax = plt.subplots(figsize=(12, 6), facecolor='white')
ax.set_facecolor('#f0f0f0') 

# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=150, ax=ax, 
                palette={"Lunch": "#7c1158", "Dinner": "#00b7c7"}, markers=["o", "X"])

# Ajustar el título y etiquetas
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20, color='black')
ax.set_xlabel('Total Bill ($)', fontsize=15, color='darkgreen')
ax.set_ylabel('Tip ($)', fontsize=15, color='darkgreen')

# Ajustar leyenda
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles', title_fontsize='13', 
          frameon=True, facecolor='white', edgecolor='black')

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.7, color='gray')

# Añadir anotación
ax.annotate('Mayor tip en la cena', xy=(45, 9), xytext=(48, 3),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12, color='black', 
            bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"))

# Añadir flecha
ax.arrow(35, 7, 10, 2, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Ajustar el espaciado
plt.tight_layout(rect=[0, 0, 1, 0.93])

# Mostrar el gráfico
plt.show()
```
