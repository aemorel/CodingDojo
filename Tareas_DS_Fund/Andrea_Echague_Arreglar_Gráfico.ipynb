```python
# CÓDIGO CON ERRORES

import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(pd.compat.StringIO(data))

# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fechas'])

# Gráfico de líneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'])
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()

# Crear un gráfico de dispersión
plt.scatter(df['fecha'], df['ventas'])
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()

# Crear un histograma de las ventas
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import io  # Para usar StringIO

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(io.StringIO(data))  # Uso de io.StringIO en lugar de pd.compat.StringIO

# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])  # Corregir el nombre de la columna 'fecha'

# Gráfico de líneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o')  # Añadir marcadores para mayor claridad
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)  # Añadir cuadrícula
plt.show()

# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red')
plt.title("Ventas Diarias - Dispersión")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
